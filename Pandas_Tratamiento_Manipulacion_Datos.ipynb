#**PROBLEMA DE NEGOCIO**


---




Se trata de un algoritmo capaz de analizar las características de una propiedad -como comodidades, tamaño, ocupación del espacio en un período determinado- y sugerir al anfitrión un precio a cobrar por tarifas diarias que garantice ganancias en momentos de alta demanda.

#**1. Entendiendo el Problema**


---




```python
import pandas as pd
```

```python
# Cargamos los datos desde un archivo JSON local
datos = pd.read_json('database/datos_hosting.json')
datos.head()
```

```python
# Normalizamos el JSON anidado
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

#**2. Datos numéricos**


---



```python
# Obtenemos los nombres de columnas
columnas = list(datos.columns)
columnas
```

```python
# Explode en multiples columnas
datos = datos.explode(columnas[3:])
datos
```

```python
# Reindexar el DataFrame después de la expansion
datos.reset_index(inplace=True, drop=True)
datos.head()
```

```python
datos.info()
```

```python
# Vemos que los datos son del tipo object, por lo que los convertimos a string
#Importamos numpy para la conversion

import numpy as np
```

```python
# Convertimos la columna 'max_hospedes' a tipo entero y la guardamos en el DataFrame
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
# Verificamos los tipos de datos nuevamente
datos.info()
```

```python
# Creamos una lista para realizar el proceso de conversion mas rapido
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']

# Convertimos las columnas numericas a tipo entero
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
datos.info()
```

```python
# Convertimos la columna 'evaluacion_general' a tipo float64
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
datos.info()
```

```python
# Removemos caracteres especiales y posibles espacios de la columna 'precio' para que no genere error al convertirlos
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos['precio']
```

```python
# Convertimos columna 'precio' a tipo float64
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
# Removemos caracteres especiales y posibles espacios de las columnas 'cuota_deposito' y 'cuota_limpieza' para que no generen error al convertirlos
# Usamos map para aplicar la funcion a ambas columnas a la vez
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].map(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
# Convertimos las columnas 'cuota_deposito' y 'cuota_limpieza' a tipo float64
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

#**3. Datos de texto**


---



```python
 datos.head()
```

```python
# Convertimos la columna 'descripcion_local' a string y minusculas
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
# Limpiamos la columna 'descripcion_local' removiendo caracteres especiales y guiones que no esten entre palabras
# Limpiamos tambien los apostrofes que no esten entre palabras
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
datos['descripcion_local'] = datos['descripcion_local'].str.replace(' (?<!\W)-(?!\w)',' ', regex=True)
```

```python
datos['descripcion_local'].head() 
```

```python
# Creacion de tokenizacion (separamos las palabras en listas)
datos['descripcion_local'] = datos['descripcion_local'].str.split()
```

```python
datos.head()
```

```python
# Limpiamos la columna 'comodidades' removiendo caracteres especiales
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"','', regex=True) 
```

```python
datos.head()
```

```python
# Creamos la tokenizacion de la columna 'comodidades' esta vez separando por comas
datos['comodidades'] = datos['comodidades'].str.split(',')
```

```python
datos.head()
```

```python
# Importamos datos de tiempo
dt_data = pd.read_json('database/inmuebles_disponibles.json')

dt_data.head()
```

```python
# Columna 'fecha' es de tipo object por lo que hay que convertirla a tipo datetime
dt_data.info()
```

```python
# Convertimos la columna 'fecha' a tipo datetime
dt_data['fecha'] = pd.to_datetime(dt_data['fecha'])
```

```python
dt_data.info() 
```

```python
# Agrupamos por año-mes
dt_data['fecha'].dt.strftime('%Y-%m')
```

```python
# Al agrupar por año-mes, procedemos a sumar los lugares disponibles por mes
# Guardamos el resultado en la variable subset
subset = dt_data.groupby(dt_data['fecha'].dt.strftime('%Y-%m'))['lugar_disponible'].sum()
subset
```
